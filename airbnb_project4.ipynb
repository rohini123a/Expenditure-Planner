```python
#identifying unique values and number of unique values in each columns for analysis

def make_hashable(value):
    if isinstance(value, list):
        return tuple(make_hashable(v) for v in value)
    elif isinstance(value, dict):
        return tuple(sorted((k, make_hashable(v)) for k, v in value.items()))
    return value

for column in df_1.columns:
    unique_values = df_1[column].apply(make_hashable).unique()
    print(f'{column} ----> {len(unique_values)}')
print("#######################################")

for column in df_1.columns:
    unique_values = df_1[column].apply(make_hashable).unique()
    if len(unique_values) <=40:
        print(f'{column} ----> {unique_values}')
        print("#######################################")
    # print(f'{column} ----> {unique_values}')
```

```python
import requests

# Your Google Maps API key
API_KEY = 'YOUR_GOOGLE_MAPS_API_KEY'

# Latitude and Longitude coordinates
latitude = -33.71472
longitude = 151.29792

# Google Maps Geocoding API URL
url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={latitude},{longitude}&key={API_KEY}"

# Send a request to the API
response = requests.get(url)
data = response.json()

# Extract Address Information
if data['status'] == 'OK':
    results = data['results']
    if results:
        for result in results:
            for component in result['address_components']:
                if 'sublocality_level_1' in component['types']:
                    suburb = component['long_name']
                elif 'locality' in component['types']:
                    city = component['long_name']
                elif 'administrative_area_level_1' in component['types']:
                    state = component['short_name']
                elif 'postal_code' in component['types']:
                    postcode = component['long_name']

        print(f"Suburb: {suburb}")
        print(f"City: {city}")
        print(f"State: {state}")
        print(f"Postcode: {postcode}")
else:
    print("Location not found.")
```
